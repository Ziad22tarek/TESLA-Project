# Importing Pakages

```python
import yfinance as yf
import pandas as pd
```

# Extracting The Data

Using the `Ticker` module we can create an object that will allow us to access functions to extract data. To do this we need to provide the ticker symbol for the stock, here the company is Tesla and the ticker symbol is `TSLA`.

```python
tsla = yf.Ticker("TSLA")
```

# Stock Information

```python
tsla_info=tsla.info
tsla_info
```

```python
# get the country
tsla_info['country']
```

# Extracting Share Prices

A share is the single smallest part of a company's stock  that you can buy, the prices of these shares fluctuate over time. Using the <code>history()</code> method we can get the share price of the stock over a certain period of time. Using the `period` parameter we can set how far back from the present to get data. The options for `period` are 1 day (1d), 5d, 1 month (1mo) , 3mo, 6mo, 1 year (1y), 2y, 5y, 10y, ytd, and max.

```python
tsla_df=tsla.history(period='max')
tsla_df
```

```python
#the types of the data
tsla_df.reset_index(inplace=True)
tsla_df.info()
```

```python
tsla_df.to_csv('TESLA.csv')
```
